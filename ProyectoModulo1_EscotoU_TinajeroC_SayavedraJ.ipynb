En esta ocasión nosotros definiremos una problemática conforme a una viga de 20 pies de longitud que se le aplica un momento en una fuerza de 15k en uno de sus extremos

Determine la pendiente y la deflexión conformela fuerza aplicada en el punto B de la viga en cantiliver por le método directo que está en la imagen: 



La fórmula general del sistema es:

$\frac{d^2}{dx^2} = \frac{M}{EI} = \frac{F}{EI} (L - x)$

La fórmula que nos permite determinar el ángulo de flexión de la viga es:

$\theta = \frac{dy}{dx} = \frac{F}{EI}(Lx-\frac{x^2}{2})$

La fórmula que nos permite determinar la flexión de la viga se obtiene integrando la fórmula anterior y está dada como:

$y=\frac{F}{EI}(\frac{L}{2}x^2 - \frac{x^3}{6})$

Y se comenzó por dividir la viga así:



Es importante denotar que en las ecuaciones anteriormente presentadas y que permiten llevar a cabo este modelo, las variables significan lo siguiente:

$F$ = Fuerza aplicada al sistema (viga)

$E$ = Módulo elástico del material 

$L$ = Longitud de la viga (en *ft*)

$I$ = Inercia del sistema

Para nuestra simulación, usaremos una viga simple de 20 *ft* de longitud total con un módulo de elasticidad de $29000(12^2)$, una inercia de $\frac{758}{12^4}$, sobre la que se aplicará una fuerza equivalente a $-15$ unidades.


## Ángulo de deflexión de la viga

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib inline

L= 20 #longitud de la viga
F= -15 #fuerza aplicada
E=29000*(12**2) #Módulo elástico del material
I=758/(12**4) #inercia
def Pendiente(y, x):
    return (F/(E*I))*((L*x)-((x*x)/2))
print(Pendiente(y, x))

x0 = 1
tt = np.linspace(0, L+1)
xx = odeint(Pendiente, x0, tt)

plt.figure(figsize=(15,10))
plt.plot(tt, xx, 'g', label = 'Ecuación de la Viga')
plt.grid()
plt.legend(loc = 'best')
plt.xlabel('Distancia $(ft)$')
plt.ylabel('Pendiente $(rad)$')
plt.title('Pendiente de la viga')
plt.show()
```

El resultado de aplicar la fórmula nos indica que la pendiente de la viga, en este caso, está dado por *-0.01965 radianes*.

## Deflexión de la viga

```python
L= 10 #longitud de la viga
F= -15 #fuerza aplicada
E=29000*(12**2) #Módulo elàstico del material
I=758/(12**4) #inercia
x=20
y=0
def Deflexión(y, x):
    return (F/(E*I))*((L*(x*x))-((x*x*x)/6))
print(Deflexión(y, x))

x0 = 1
tt = np.linspace(0, L+1)
xx = odeint(Deflexión, x0, tt)

plt.figure(figsize=(15,10))
plt.plot(tt, xx, 'g', label = 'Ecuación de la Viga')
plt.grid()
plt.legend(loc = 'best')
plt.xlabel('Distancia $(ft)$')
plt.ylabel('Deflexión')
plt.title('Deflexión de una viga')
plt.show()
```

El resultado de aplicar la fórmula nos indica que la deflexión está dada, en este caso, en *-0.262 ft.*
